**🔥 What is np.random.seed()?**
* 👉 It fixes randomness
* 👉 So you get the same random numbers every time
**🔥 1. What is Random in Computers?**
* 👉 Computers cannot generate true randomness ❌
* 👉 They generate pseudo-random numbers ✔️
* 🔹 Meaning
* 👉 Numbers look random
* 👉 But actually follow a deterministic formula (algorithm)
**🔥 2. What is a Random Generator?*8
* 👉 NumPy uses a Pseudo Random Number Generator (PRNG)
* 👉 It works like:
**next_number = function(previous_number)**
* 👉 So:
* 👉 Every number depends on the previous one
**🔥 3. What is SEED?**
* 👉 Seed = initial value (starting number) given to PRNG
* 🔹 Mathematical idea
* x₀ = seed
* x₁ = f(x₀)
* x₂ = f(x₁)
* x₃ = f(x₂)
* ...
**👉 This sequence is fixed once seed is fixed**
* 🔥 4. Example (Conceptual)
* Case 1:
* np.random.seed(42)
* np.random.randint(1,10,3)
* 👉 Output:
* [7 4 8]
* Case 2 (same seed):
* np.random.seed(42)
* np.random.randint(1,10,3)
* 👉 Output:
* [7 4 8]   ✔️ same
* Case 3 (different seed):
* np.random.seed(10)
* 👉 Output changes ❗
**🔥 5. Important Properties**
* ✅ Deterministic
👉 Same seed → same sequence
* ✅ Reproducibility
* 👉 Very important in:
* Machine Learning
* Research
* Experiments
* ❌ Not truly random
* 👉 If seed is known → output can be predicted

* 🔹 2. flatten() and ravel()
* 🔥 Meaning
* 👉 Convert multi-dimensional array → 1D array
* 🔥 Difference
* Function	Behavior
* flatten()	creates a copy
* ravel()	creates a view
* 👉 Simple:
* copy = new memory
* view = same memory
* flatten() = independent copy
* ravel() = linked to original

```python
arr = np.array([[1,2,3],[4,5,6]])
f = arr.flatten()
f
```

```python
f[0] = 100
print(arr)
print(f)
```

* 👉 Now:
* f is a separate new array
* data is copied
* 👉 Original array NOT changed ✔️
* 🔹 Why?
* 👉 Because:
* 👉 flatten() made a copy (new memory)

```python
arr = np.array([[1,2,3],[4,5,6]])
r = arr.ravel()
r
```

* 👉 Now:
* r is linked to original array
* same memory

```python
r[0] = 100
print(r)
print(arr)
```

**👉 Original array changed ❗**
* 🔥 Why?
* 👉 Because:
* 👉 ravel() did NOT create new memory
* 👉 It just created a view (reference)

🔥 Core Difference (simple)

👉 concatenate = join arrays along an existing axis
👉 stack = join arrays by creating a new axis

🔹 1. np.concatenate()

👉 Joins arrays without changing dimensions

Example
a = np.array([1,2])
b = np.array([3,4])

np.concatenate((a,b))

👉 Output:

[1 2 3 4]

👉 Shape:

(4,)
2D Example
a = [[1,2],
     [3,4]]

b = [[5,6],
     [7,8]]

np.concatenate((a,b), axis=0)

👉 Output:

[[1 2]
 [3 4]
 [5 6]
 [7 8]]

👉 Rows added ✔️

🔹 2. np.stack()

👉 Joins arrays by adding a new dimension

Example
a = np.array([1,2])
b = np.array([3,4])

np.stack((a,b))

👉 Output:

[[1 2]
 [3 4]]

👉 Shape:

(2,2)
🔥 Key Visualization
👉 concatenate
[1 2] + [3 4]
→ [1 2 3 4]

👉 Same dimension

👉 stack
[1 2]
[3 4]

👉 New dimension added

🔥 Axis understanding
concatenate
np.concatenate((a,b), axis=0)

👉 existing axis use hota hai

stack
np.stack((a,b), axis=0)

👉 new axis insert hota hai

🔴 Important Rule
Feature	concatenate	stack
Axis	existing	new
Shape change	no	yes
Dimension	same	increases
🔥 Memory

👉 Both create COPY (new memory) ✔️

🧠 Easy Trick

👉
